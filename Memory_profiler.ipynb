```python
from functools import wraps
from time import sleep
```

```python
def memory_profile(fn):
    import psutil
    @wraps(fn)
    def inner(*args,**kwargs):
        before = psutil.virtual_memory().available / (1024.0 ** 2)
        result = fn(*args,**kwargs)
        after = before - psutil.virtual_memory().available / (1024.0 ** 2)
        print(f"Memory consumed is {before - after} for {fn.__name__}")
        return result

    return inner
```

```python
@memory_profile
def f():
    # a function that with growing
    # memory consumption
   

    a = [0] * 1000
    sleep(.1)
    b = a * 100
    sleep(.1)
    c = b * 100
    
    return a
```
